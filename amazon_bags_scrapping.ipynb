```python
from bs4 import BeautifulSoup
import requests
import csv
```

```python
import pandas as pd
```

```python
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:66.0) Gecko/20100101 Firefox/66.0", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

source = requests.get('https://www.amazon.in/s?k=bags&crid=2M096C6104MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_1', headers = headers).text

soup = BeautifulSoup(source, 'lxml')

print(soup.prettify())
```

```python
Names =[]
Prices=[]
Reviews=[]
Ratings=[]
```

```python
for i in soup.find_all('a', class_='a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'):
  string = i.text
  Names.append(string.strip() )
for each in Names:
  sort = False
  print(each)  
```

```python
for i in soup.find_all('span',class_='a-price-whole'):
  string = i.text
  Prices.append(string.strip() )
for each in Prices:
  #sort = False
  print(each)  
  #string = i.text
  #Prices.append(string.strip())  
#for each in Prices:
  #print(each)  
  
  
  
  
```

```python
for i in soup.find_all('span', class_='a-icon-alt'):
  string = i.text
  Reviews.append(string.strip() )
  #print(Reviews)
for each in Reviews:
  print(each)

  
```

```python
for i in soup.find_all('span', class_='a-size-base s-underline-text'):
  string = i.text
  Ratings.append(string.strip() )
for each in Ratings:
  print(each)  
  
  
  
  
```

```python
file_name = 'amazon_scraping'

with open(file_name, 'w') as file:
  writer = csv.writer(file)
  writer.writerow(['Bags Name','Prices', 'Reviews', 'Ratings'])

  for i in range(len(Names)):
    writer.writerow([i, Names[i], Prices[i], Reviews[i], Ratings[i]])
```
